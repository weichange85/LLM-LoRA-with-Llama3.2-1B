```python
import json
import pandas as pd
import pathlib as Path
import os
```

```python
# load data
with open("../data/train.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
print(f"Total samples: {len(data)}")
```

```python
# basic preview
print("\nExample entry:")
print(data[0])
```

```python
# extract prompts and responses
prompts = [x["prompt"] for x in data]
responses = [x["response"] for x in data]
```

```python
# length analysis (word count)
prompt_lengths = [len(p.split()) for p in prompts]
response_lengths = [len(r.split()) for r in responses]
```

```python
print("\nPrompt length stats (words)")
print(f"min: {min(prompt_lengths)}")
print(f"max: {max(prompt_lengths)}")
print(f"avg: {sum(prompt_lengths)/len(prompt_lengths):.2f}")

print("\nResponse length stats (words)")
print(f"min: {min(response_lengths)}")
print(f"max: {max(response_lengths)}")
print(f"avg: {sum(response_lengths)/len(response_lengths):.2f}")
```
